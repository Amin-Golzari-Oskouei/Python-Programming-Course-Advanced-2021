```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from numpy.random import randn
np.random.seed(45)
```

```python
m = ['movie_id', 'title', 'genres']
movies = pd.read_table('data/movies.dat', sep='::', header=None, names=m)
```

```python
u = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('data/users.dat', sep='::', header=None, names=u)
```

```python
r = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('data/ratings.dat', sep='::', header=None, names=r)
```

```python
df = pd.merge(pd.merge(ratings, users), movies)
df
```

```python
df.iloc[0]
```

```python
mr = df.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mr[:10]   # mr : mean_ratings
```

```python
rt = df.groupby('title').size()     # rt : ratings_by_title
rt[:10]
```

```python
at = rt.index[rt >= 250]   # at :active_titles
at
```

```python
mr = mr.loc[at]
mr
```

```python
tm = mr.sort_values(by='M', ascending=False)
tm[:15]                                          # tm : top_male_ratings
```

```python
tf = mr.sort_values(by='F', ascending=False)
tf[:15]                                          # tf : top_female_ratings
```

### Measuring Rating Disagreement

```python
mr['diff'] = mr['F'] - mr['M']
mr.head(10)
```

```python
sbd = mr.sort_values(by='diff')
sbd[:10]                                  # sbd : sorted_by_diff
```

```python
sbd[::-1][:5]
```
