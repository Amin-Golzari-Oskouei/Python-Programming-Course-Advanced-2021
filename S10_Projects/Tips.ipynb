```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from numpy.random import randn
import random  
np.random.seed(10)
```

```python
dftips = pd.read_csv('data/tips.csv')
dftips
```

```python
sns.boxplot(x="time", y="tip", data=dftips);
```

```python
sns.boxplot(x="time", y="tip", data=dftips, order=["Lunch", "Dinner"]);
```

```python
sns.boxplot(x="day", y="total_bill", data=dftips);
```

```python
sns.boxplot(x="day", y="total_bill", data=dftips);
sns.swarmplot(x="day", y="total_bill", data=dftips, color=".25", linewidth=2.5);
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=dftips, palette="Set1");
```
