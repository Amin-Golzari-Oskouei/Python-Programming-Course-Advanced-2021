## Merging Datasets

```python
df1 = pd.DataFrame({'Name': ['ali', 'taha', 'sara'], 'C++': [18, 12, 17]})
df1
```

```python
df2 = pd.DataFrame({'Name': ['ali', 'taha', 'omid'], 'Python': [13, 14, 16]})
df2
```

```python
dfm = df1.merge(df2, on='Name')
dfm
```

```python
df1.merge(df2, how='inner', on='Name')
```

```python
df1.merge(df2, how='outer', on='Name')
```

```python
df1.merge(df2, how='left', on='Name')
```

```python
df1.merge(df2, how='right', on='Name')
```

### Merging on Index

```python
df1 = pd.DataFrame({'X': ['a', 'b', 'a', 'a', 'b', 'c'], 'Y': range(6)})
df1
```

```python
df2 = pd.DataFrame({'Z': [18, 15]}, index=['a', 'b'])
df2
```

```python
pd.merge(df1, df2, left_on='X', right_index=True)
```

```python
pd.merge(df1, df2, left_on='X', right_index=True, how='outer')
```

```python
####
```

```python
d = {'X': ['H', 'H', 'H', 'N', 'N'],'Y': [1397, 1398, 1399, 1398, 1399], 'Z': np.arange(5.)}
frame1 = pd.DataFrame(d)
frame1
```

```python
i = [['N', 'N', 'H', 'H', 'H', 'H'], [1398, 1397, 1397, 1397, 1398, 1399]]

frame2 = pd.DataFrame(np.arange(12).reshape((6, 2)), index=i, columns=['A', 'B'])
frame2
```

```python
pd.merge(frame1, frame2, left_on=['X', 'Y'], right_index=True)
```

```python
pd.merge(frame1, frame2, left_on=['X', 'Y'], right_index=True, how='outer')
```

# pivot

```python
df = pd.DataFrame({
       "A": ['one', 'one', 'one', 'two', 'two', 'two'],
       "B": [1, 1, 2, 1, 1, 2],
       "C": [1, 2, 1, 2, 1, 2],
       "V": [0, 1, 2, 3, 4, 5]})
df
```

```python
df.pivot(index="A", columns=["B", "C"],values="V")
```

```python
df.pivot(index=["A", "B"], columns=["C"],values="V")
```

# melt

```python
mydict = {'Name': {0:'Ali', 1: 'Sara', 2: 'Mahsa'}, 'C++': {0: 12, 1: 13, 2: 20}, 'Python': {0: 14, 1: 16, 2: 8}}
df = pd.DataFrame(mydict)
df
```

```python
df.melt(['Name'])                    # df.melt(id_vars=['Name'])
```

```python
m = pd.melt(df, ['Name'])
r = m.pivot('Name', 'variable', 'value')
r
```

```python
r.reset_index()
```
